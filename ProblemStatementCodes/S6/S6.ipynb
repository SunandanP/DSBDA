Perform the following operations using Python on the Facebook metrics data sets

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Amazonbookreviews.csv")
df
```

```python
# /help drop rows containing the missing values
df = df.iloc[:, :-2]
df
```

```python
columns = df.columns
```

```python
columns
```

```python
subsets = []
size = 4
for i in range(0, len(columns), size):
    if i <= len(columns):
        subsets.append(df.iloc[:, i: i+size])
subsets
```

```python
subsets[0]
```

```python
subsets[1]
```

# Merging datasubsets


```python
mergable_dataset = []
mergable_dataset.append(df.iloc[:, 0:3])
mergable_dataset.append(df.iloc[:, 2:])
```

```python
mergable_dataset
```

```python
# Merging of column based subsets
```

```python
merged_dataset = mergable_dataset[0].merge(mergable_dataset[1], on="reviewername", how="inner")
merged_dataset
```
